```python
df=pd.read_csv('submission_with_punctuation.csv')
```

```python
df['punctuation']=df['punctuation'].str[:-1]
```

```python
df['punctuation'][1575]
```

```python
df['punctuation'] = df['punctuation'].str.replace('।', '')
```

```python
df['punctuation'] = df['punctuation'].str.replace('.', '')
```

```python
df
```

```python
df=df.drop('sentence',axis=1)
df['sentence'] = df['punctuation']
df=df.drop('punctuation',axis=1)
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
comma_id=df.index[df['sentence'].str[-1] == ',']
```

```python
nan_id=df.loc[pd.isna(df["sentence"]), :].index
```

```python
deleted = df[df.index.isin(nan_id)]['path'].tolist()
```

```python
df=df.drop(nan_id)
df=df.reset_index(drop=True)
```

```python
df
```

```python
for i in range(len(df)):
    if df['sentence'][i][-1]==',':
        df['sentence'][i] = df['sentence'][i][:-1]
        print(df['sentence'][i])
```

```python
for i in range(len(df)):
    if df['sentence'][i][-1] not in [',','?']:
        df['sentence'][i] = df['sentence'][i] + '।'
```

```python
df
```

```python
for i in range(len(deleted)):
    df = df.append({'path': deleted[i],'sentence':'',},ignore_index=True)
```
